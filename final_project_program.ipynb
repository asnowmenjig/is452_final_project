```python
#VOLUME 1
# Dictionary content starts on line 7245
# A = lines 7245 to 33259 (26014 lines)
# B = lines 33260 to 56783 (23523 lines)
# C = lines 56784 to 98775 (41992 lines)
# D = lines 98776 to 124747 (25972 lines)
# E = lines 124748 to 141819 (17071 lines)
# F = lines 141820 to 165059 (23240 lines)
# G = lines 165060 to 179640 (14581 lines)
# H = lines 179641 to 194989 (15349 lines)
# IJ = lines 194990 to 215940 (20951 lines)
# K = lines 215941 to 218433 (2493 lines)

#VOLUME 2
# Dictionary content starts on line 85
# L = lines 85 to 15134             #page np-57
# M = lines 15137 to 35303          #page 58-162
# N = lines 35304 to 42147          #page 163-199
# O = lines 42148 to 52527          #page 200-255
# P = lines 52528 to 86702          #page 256-256
# Q = lines 86702 to 89231
# R = lines 89231 to 110052         #page 256-287
# S = lines 110053 to 163742        #page 288-np
# T = lines 163742 to 184958        #page np-np
# V/U = lines 184958 to 207127
# W = lines 207128 to 221750
# Y = lines 221750 to 222862
# Z = lines 222863 to 223136

#############################################

## CLEANING ##                                      #converts text to lowercase, cleans up punctuation and
                                                    # special characters, then outputs cleaned text
import string

def clean_line(line_str):
    cleanline = line_str.lower()
    for punc in string.punctuation:
         cleanline = cleanline.replace(punc, "")
    cleanline = cleanline.replace("", "o")
    cleanline = cleanline.replace("", "o")
    cleanline = cleanline.replace("", "x")
    cleanline = cleanline.replace("", "a")
    cleanline = cleanline.replace("", 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'a')
    cleanline = cleanline.replace('', 'c')
    cleanline = cleanline.replace('', 'c')
    cleanline = cleanline.replace('', 'c')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'e')
    cleanline = cleanline.replace('', 'g')
    cleanline = cleanline.replace('', 'i')
    cleanline = cleanline.replace('', 'i')
    cleanline = cleanline.replace('', 'n')
    cleanline = cleanline.replace('', 'n')
    cleanline = cleanline.replace('', 'n')
    cleanline = cleanline.replace('', 'o')
    cleanline = cleanline.replace('', 'o')
    cleanline = cleanline.replace('', 'o')
    cleanline = cleanline.replace('', 'o')
    cleanline = cleanline.replace('', 'o')
    cleanline = cleanline.replace('', 'r')
    cleanline = cleanline.replace('', 's')
    cleanline = cleanline.replace("", "s")
    cleanline = cleanline.replace('', 's')
    cleanline = cleanline.replace('', 's')
    cleanline = cleanline.replace('', 't')
    cleanline = cleanline.replace('', 'u')
    cleanline = cleanline.replace('', 'u')
    cleanline = cleanline.replace('', 'u')
    cleanline = cleanline.replace('', 'z')
    cleanline = cleanline.replace('', 'z')
    cleanline = cleanline.replace('', 'z')
    return cleanline

##########################################

## LIST OF WORDS ##

infile = open('johnson_v1_A.txt','rb')      #read alphabet section file
text = str(infile.read())
infile.close()

lines_list = text.split('\n')                                 #split text into list of lines
#print(lines_list)

for line in lines_list:                                       #clean list of lines
    clean = clean_line(line)
#    print(clean)
    clean_words = clean.split()                              #split cleaned lines into list of cleaned words
# print(clean_words)

###########################################

## DICTIONARY COUNT ##                                        #adds words to dictionary and keeps count of frequency

word_counts = dict()                                          #create dictionary for word counts
for word in clean_words:
    if word not in word_counts:
        word_counts[word] = 1
    else:
        word_counts[word] += 1
# print(word_counts)

###########################################

## CSV COUNT ##                                     #writes dictionary (word_counts) contents to CSV file

import csv

outfile = open('wordcounts_v1_A.csv', 'w', newline = '', encoding="UTF-8")      ######################
csvout = csv.writer(outfile)
csvout.writerow(['word', 'count'])

for pair in word_counts.items():
    label = pair[0]
    count = pair[1]
    row = [label, count]
    csvout.writerow(row)
```
